```python
import ast
def convert(obj):  #ye ek string hai
    L=[]
    for i in ast.literal_eval(obj):  #yaha pr iss operation k baad wo list ban jayega
        L.append(i['name'])
    return L    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

## now we will repeat the same process for other colums also

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):  #ye ek string hai
    L=[]
    counter=0
    for i in ast.literal_eval(obj):  #yaha pr iss operation k baad wo list ban jayega
       if counter !=3: 
         L.append(i['name'])
         counter+=1
       else: 
          break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

## now we have to convert for crew but we choose only director

```python
def fetch_director(obj):  #ye ek string hai
    L=[]
    for i in ast.literal_eval(obj): 
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L    
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

## now we will convert this overview column into list so that we can merge it with other lists

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

## now we will remove the spaces between the words to make it a single entity

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

## now we will make a new column called "tags" and concatenate all thse column into it

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

## so now we use vectorization
means coverting the text into vector
we use bag of words method here : in this we make avery large string by concatenating all the text(tags) and take out most used words
then we don't consider "stop words" : these are the words that used in english for sentence formation but has no contribution in
sentence meaning.eg-are,and,if,to,from etc.
to do this we use a library in scikit learn called "count Vector" now,

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #here we explicitely convery into numpy array.
#by default CountVector returns a scipi spars matrix so we convert it into numpy array(which we need here)
```

## now we apply steming to remove the duplicate words 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):    #we are creating a helper function
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)    #converting back to string
```

```python
['loved','loving','love']
['love','love','love']
```

```python
ps.stem('dancing')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## now we will use cosine distance to find similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]  # fetching the index of a movie
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Batman Begins')
```
